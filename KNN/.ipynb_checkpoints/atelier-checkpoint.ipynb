**5 - Quand utiliser la normalisation et quand utiliser la standardisation :**<br>
<li>La normalisation est bonne à utiliser lorsque vous savez que la distribution de vos données ne suit pas une distribution gaussienne. Cela peut être utile dans les algorithmes qui ne supposent aucune distribution des données comme K-Nearest Neighbours et Neural Networks.</li>
<li>La Standardization, peut être utile dans les cas où les données suivent une distribution gaussienne. Cependant, cela ne doit pas nécessairement être vrai. De plus, contrairement à la normalisation, la normalisation n'a pas de plage de délimitation. Ainsi, même si vous avez des valeurs aberrantes dans vos données, elles ne seront pas affectées par la normalisation.</li>

**6 - Ecrire une fonction qui permet de faire la standardisation et une autre fonction pour faire la normalisation :**<br>

**7 -  Standardiser les données du tableau :**<br>


```python
df = pd.read_csv('Iris.csv')
df.head()
```

```python
def normalisation(X):
    return [(x - X.min())/(X.max() - X.min()) for x in X]
normalisation(df[:10]['SepalLengthCm'])
```

```python
def standardisation(X):
    moyen = np.average(X)
    std = np.std(X)
    return [(x - moyen)/std for x in X]
standardisation(df[:10]['SepalLengthCm'])
```

**8 -  En utilisant la distance euclidienne, calculer les distances de x aux autres enregistrements. (vous pouvez
implémenter une méthode qui calcule la distance entre un ensemble et un nouvel enregistrement et
retourner un vecteur de distance) :**<br>

```python
def euclidienne_numpy(X, y):
    return [np.linalg.norm(x-y) for x in X]
euclidienne_numpy(df[:10]['SepalLengthCm'],df.iloc[0]['SepalLengthCm'])
```

```python
def euclidienne(X, y):
    return [np.sqrt(np.sum(np.power(np.abs(x-y), 2)))for x in X]
euclidienne(np.array(df.iloc[:,:-1]),np.array(df.iloc[1,:-1]))
```

### <font color='red'>5 - Exercice 4 : Evaluation du modèle en fonction de k</font> ###

**1 - Répartir la base de données en dataset_learn (80%) et dataset_test (20%)**<br>

```python
df['label'] = df['Species'].cat.codes
df.head()
```

```python
X = df[['SepalLengthCm', 'SepalWidthCm', 'PetalLengthCm','PetalWidthCm']]
y = df['label']

# default is 80% / 20% train-test split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state=0)
```

**2 - Créer un classifier en précisant les paramètres qu’il faut**<br>

**3 - Mesurer accuracy du classifier**

```python
from sklearn.neighbors import KNeighborsClassifier
def Classifier(X, y, k):
    classifier= KNeighborsClassifier(n_neighbors=k)
    classifier.fit(X, y)
    y_pred = classifier.predict(X)
    accuracy = accuracy_score(y_pred, y)
    return accuracy
```

**4- Ecrire le code source pour mesurer la accuracy pour k allant de 3 à 20**

```python
def plot_accuracy(X_train, y_train):
    accuracies = []
    for k in range(1,20):
        accuracies.append(Classifier(X_train, y_train, k))
    plt.figure(figsize=(12,8))
    plt.plot(range(1,20),accuracies,color='blue',linestyle='dashed', marker='o',markerfacecolor='red',markersize=10)
    plt.title('Accuracy vs K value')
    plt.xlabel('K')
    plt.ylabel('Accuracy')
    print('best accuracy :',max(accuracies))
```

```python
plot_accuracy(X_train, y_train)
```

```python
X = df_copy[['SepalLengthCm', 'SepalWidthCm', 'PetalLengthCm','PetalWidthCm']]
y = df_copy['label']

# default is 80% / 20% train-test split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state=0)
```

```python
plot_accuracy(X_train, y_train)
```
