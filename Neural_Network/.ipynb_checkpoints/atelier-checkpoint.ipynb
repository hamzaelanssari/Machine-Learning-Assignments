### 4 - Exercice 3 : Xor Problem ###

1 - Analyser le code source.

```python
#Program2:
from sklearn.svm import SVC
import matplotlib.pyplot as plt
from mlxtend.plotting import plot_decision_regions
import numpy as np
rng = np.random.RandomState(0)
X = rng.randn(300, 2)
y = np.array(np.logical_xor(X[:, 0] > 0, X[:, 1] > 0), dtype=int)
svm_model = SVC(gamma='auto')
svm_model.fit(X, y)
fig = plt.figure(figsize=(10,8))
fig = plot_decision_regions(X=X, y=y, clf=svm_model, legend=2)
plt.title("SVM")
plt.show()
```

```python
X.shape[1]
```

2 - Créer un réseau de neurone ayant trois couches (le nombre de neurones de la couche cachée est 2)

```python
model_3 = Sequential(name="XOR_Problem")
hidden_layer=Dense(2, input_shape=[X.shape[1]], activation="relu",name="Hidden_Layer")
model_3.add(hidden_layer)
output_layer=Dense(1, activation="sigmoid", name="Output_Layer")
model_3.add(output_layer)
```

```python
model_3.summary()
```

```python
model_3.get_weights()
```

```python
model_3.compile(
    loss="binary_crossentropy",
    optimizer="adam",
    metrics=["accuracy","mse"])
model_3.fit(X, y,epochs=1000)
```

```python
metrics=model_3.evaluate(X,y)
print(metrics)
```

```python
fig = plt.figure(figsize=(10,8))
fig = plot_decision_regions(X=X, y=y, clf=model_3, legend=2)
plt.title("XOR Multilayer NN Model")
plt.show()
```
