1 - À l’aide du programme #program1 de l’annexe de l’atelier précédent, générer 100 enregistrements.

```python
def f(x):
    return -x + 5


def generateData(n):
    # generate random numbers between 0-1
    seed(1)
    min = 0
    max = 5
    dataset = []
    for _ in range(n):
        valueX1 = random()
        scaledvalueX1 = min + (valueX1 * (max - min))
        valueX2 = random()
        scaledvalueX2 = min + (valueX2 * (max - min))
        x1 = round(scaledvalueX1, 1)
        x2 = round(scaledvalueX2, 1)
        y0 = f(x1)
        cl = 0
        if (x2 > y0):
            cl = 1
        dataset.append([x1, x2, cl])
    return dataset

dataset = np.array(generateData(100))
dataset
```

2 - Tracer les données dans un graphe

```python
def plotData(dataset):
    c0 = dataset[:, 2] == 0
    c1 = dataset[:, 2] == 1
    XD1 = dataset[:, 0][c0]
    YD1 = dataset[:, 1][c0]
    XD2 = dataset[:, 0][c1]
    YD2 = dataset[:, 1][c1]
    plt.plot(XD1, YD1, 'rD')
    plt.plot(XD2, YD2, 'bo')
    plt.show()

plotData(dataset)
```

3. Modifier le code source de manière à afficher les éléments suivants à chaque 10000 itérations
<br>a. Accuracy
<br>b. Loss value (binary_crossentropy)

```python
def predict(bias,W,X):
    linear_model=np.dot(X,W.T)+bias
    y_predicted=sigmoid(linear_model)
    y_predicted_cls=[1 if i>0.5 else 0 for i in y_predicted]
    return y_predicted_cls, y_predicted

def sigmoid(x):
    return 1/(1+np.exp(-x))

def binary_cross_entropy(y, y_propa):
    cross_entropy = np.dot(y, np.log(y_propa)) +np.dot((1-y), np.log(1-y_propa))
    cost = -np.sum(cross_entropy)/float(len(y_propa))
    return cost

def g(bias, W, x):
    return (-bias/W[1])-(W[0]/W[1])*x
```

```python
lr=0.001
n_iters=100000
c0 = dataset[:, 2] == 0
c1 = dataset[:, 2] == 1
XD1 = dataset[:, 0][c0]
YD1 = dataset[:, 1][c0]
XD2 = dataset[:, 0][c1]
YD2 = dataset[:, 1][c1]
X = dataset[:,:2]
y = dataset[:,2]
y = y.astype(int)
list1 = list(zip(XD1,YD1))
list2 = list(zip(XD2,YD2))
def fit(X,y):
    bias=0
    W=np.array([0.,0.])
    n_samples,n_features=X.shape
    for i in range(n_iters):
        y_predicted, y_proba=predict(bias, W, X)
        dW=(1/n_samples)*np.dot(X.T,(y_proba-y))
        dbias=(1/n_samples)*np.sum(y_proba-y)
        W-=lr*dW #ligne12
        bias-=lr*dbias #ligne13
        if i%10000==0:
            print('-------------------iteration ',i,'-------------------')
            print('Bias : ',bias)
            print('Weights : ',W)
            print('Accuracy : ',accuracy_score(y_predicted,y))
            print('Binary Cross Entropy : ',binary_cross_entropy(y,y_proba))
            plt.plot(XD1, YD1, 'gD')
            plt.plot(XD2, YD2, 'bo')
            plt.plot([0,g(bias, W, 0)],[5,g(bias, W, 5)])
            line_vector = (5,g(bias, W, 5)-g(bias, W, 0))
            XD1_error, YD1_error, XD2_error, YD2_error= ([] for i in range(4))
            for i in list1:
                point_vector = (5-i[0],g(bias, W, 5)-i[1])
                xp = line_vector[0]*point_vector[1] - line_vector[1]*point_vector[0]  # Cross product
                if xp<0:
                    XD1_error.append(i[0])
                    YD1_error.append(i[1])
            for i in list2:
                point_vector = (5-i[0],g(bias, W, 5)-i[1])
                xp = line_vector[0]*point_vector[1] - line_vector[1]*point_vector[0]  # Cross product
                if xp>0:
                    XD2_error.append(i[0])
                    YD2_error.append(i[1])
            plt.plot(XD1_error, YD1_error, 'rx')
            plt.plot(XD2_error, YD2_error, 'rx')
            plt.show()
    return bias,W
b,w=fit(X,y)
```
